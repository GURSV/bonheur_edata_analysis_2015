```python
fig = px.scatter(df, x='Country', y='Happiness Score', title='Country Vs Happiness')
fig.show()
```

```python
fig = px.scatter(df, x='Country', y='Happiness Rank', title='Country Vs Happiness')
fig.show()
```

```python
fig = px.scatter(df, x='Country', y='Economy (GDP per Capita)', title='Country Vs Economy')
fig.show()
```

```python
fig = px.scatter(df, x='Economy (GDP per Capita)', y='Happiness Rank', title='Economy Vs Happiness')
fig.show()
```

```python
fig = px.scatter(df, x='Happiness Rank', y='Family', title='Happiness Vs Family')
fig.show()
```

```python
fig = px.scatter(df, x='Country', y='Family', title='Country Vs Family')
fig.show()
```

```python
# plt.figure(figsize=(10, 10))
sns.pairplot(df[['Happiness Rank', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']])
```

```python
corr = df[['Happiness Score', 'Health (Life Expectancy)', 'Economy (GDP per Capita)', 'Family', 'Freedom', 'Trust (Government Corruption)', 'Generosity']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.jointplot(x='Happiness Rank', y='Dystopia Residual', data=df, kind='hex')
plt.show()
```

The plots clearly show how different parameters are related to each other. Relation/Impact of each parameter with/on Happiness is described through these plots.
